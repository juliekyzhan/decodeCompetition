#### employ_industry_dist

```python
# 创建新列用于存储编辑距离结果
df_cash['employ_industry_dist'] = 0

# 处理可能的缺失值，用空字符串填充
for col in ['employ1', 'employ2', 'employ3', 'industry_tp1', 'industry_tp2', 'industry_tp3']:
    df_cash[col] = df_cash[col].fillna('').astype(str)

# 对每一行计算编辑距离
for idx, row in df_cash.iterrows():
    # 构建映射字典，将所有不同的值映射到不同字符
    unique_values = set([
        row['employ1'], row['employ2'], row['employ3'],
        row['industry_tp1'], row['industry_tp2'], row['industry_tp3']
    ])
    
    # 创建映射字典，相同的值映射到相同的字母
    value_to_char = {val: chr(97 + i) for i, val in enumerate(unique_values)}
    
    # 映射值到字符
    employ_str = value_to_char[row['employ1']] + value_to_char[row['employ2']] + value_to_char[row['employ3']]
    industry_str = value_to_char[row['industry_tp1']] + value_to_char[row['industry_tp2']] + value_to_char[row['industry_tp3']]
    
    # 计算编辑距离
    dist = Levenshtein.distance(employ_str, industry_str)
    
    # 存储结果
    df_cash.at[idx, 'employ_industry_dist'] = dist

# 查看结果分布
print(df_cash['employ_industry_dist'].describe())
print(df_cash['employ_industry_dist'].value_counts())

# # 可以根据距离创建正负特征
# df_cash['employ_industry_close'] = (df_cash['employ_industry_dist'] <= 1).astype(int)  # 0或1表示匹配度高
# df_cash['employ_industry_far'] = (df_cash['employ_industry_dist'] >= 2).astype(int)   # 2或3表示匹配度低

# # 显示结果
# print("\n匹配度高的样本数(0-1):", df_cash['employ_industry_close'].sum())
# print("匹配度低的样本数(2-3):", df_cash['employ_industry_far'].sum())
```

#### ethnicity_dist

```python
# 创建新列用于存储编辑距离结果
df_cash['ethnicity_dist'] = 0

# 处理可能的缺失值，用空字符串填充
for col in ['ethnicity1', 'ethnicity2', 'ethnicity3', 'ethnicity_tp1', 'ethnicity_tp2', 'ethnicity_tp3']:
    df_cash[col] = df_cash[col].fillna('').astype(str)

# 对每一行计算编辑距离
for idx, row in df_cash.iterrows():
    # 构建映射字典，将所有不同的值映射到不同字符
    unique_values = set([
        row['ethnicity1'], row['ethnicity2'], row['ethnicity3'],
        row['ethnicity_tp1'], row['ethnicity_tp2'], row['ethnicity_tp3']
    ])
    
    # 创建映射字典，相同的值映射到相同的字母
    value_to_char = {val: chr(97 + i) for i, val in enumerate(unique_values)}
    
    # 映射值到字符
    ethnicity_str = value_to_char[row['ethnicity1']] + value_to_char[row['ethnicity2']] + value_to_char[row['ethnicity3']]
    ethnicity_tp_str = value_to_char[row['ethnicity_tp1']] + value_to_char[row['ethnicity_tp2']] + value_to_char[row['ethnicity_tp3']]
    
    # 计算编辑距离
    dist = Levenshtein.distance(ethnicity_str, ethnicity_tp_str)
    
    # 存储结果
    df_cash.at[idx, 'ethnicity_dist'] = dist

# 查看结果分布
print(df_cash['ethnicity_dist'].describe())
print(df_cash['ethnicity_dist'].value_counts())

# # 可以根据距离创建正负特征
# df_cash['ethnicity_close'] = (df_cash['ethnicity_dist'] <= 1).astype(int)  # 0或1表示匹配度高
# df_cash['ethnicity_far'] = (df_cash['ethnicity_dist'] >= 2).astype(int)   # 2或3表示匹配度低

# # 显示结果
# print("\n匹配度高的样本数(0-1):", df_cash['ethnicity_close'].sum())
# print("匹配度低的样本数(2-3):", df_cash['ethnicity_far'].sum())
```

```python
# 0602版本更新（修改后）

# 1. 使用z-score标准化处理party_tp_score和npvi变量
from scipy import stats
import numpy as np

# 对party_tp_score标准化，先处理缺失值
party_tp_score_filled = df_cash['party_tp_score'].fillna(0)  # 临时填充为0
df_cash['party_tp_score_z'] = stats.zscore(party_tp_score_filled)

# 对npvi标准化，先处理缺失值
npvi_filled = df_cash['npvi'].fillna(0)  # 临时填充为0
df_cash['npvi_z'] = stats.zscore(npvi_filled)

# 2. 计算两个标准化变量的差的绝对值
df_cash['score_diff_abs'] = abs(df_cash['party_tp_score_z'] - df_cash['npvi_z'])

# 3. 计算绝对差值的中位数（作为划分阈值）
diff_median = df_cash['score_diff_abs'].median()
print(f"两个标准化变量绝对差值的中位数: {diff_median}")

# 4. 初始化party_score_pl和party_score_ne变量
df_cash['party_score_pl'] = 0.0
df_cash['party_score_ne'] = 0.0

# 5. 根据绝对差值与中位数的关系设置变量值
# 绝对差值小于或等于中位数：pl=绝对差值，ne=0
mask_le_median = df_cash['score_diff_abs'] <= diff_median
df_cash.loc[mask_le_median, 'party_score_pl'] = df_cash.loc[mask_le_median, 'score_diff_abs']
df_cash.loc[mask_le_median, 'party_score_ne'] = 0

# 绝对差值大于中位数：ne=绝对差值，pl=0
mask_gt_median = df_cash['score_diff_abs'] > diff_median
df_cash.loc[mask_gt_median, 'party_score_ne'] = df_cash.loc[mask_gt_median, 'score_diff_abs']
df_cash.loc[mask_gt_median, 'party_score_pl'] = 0

# 显示结果
print("\nparty_score特征构造结果:")
print(f"绝对差值≤中位数的样本数: {mask_le_median.sum()}")
print(f"绝对差值>中位数的样本数: {mask_gt_median.sum()}")

print("\nparty_score_pl描述统计:")
print(df_cash['party_score_pl'].describe())
print("\nparty_score_ne描述统计:")
print(df_cash['party_score_ne'].describe())

# # 查看前10行结果
# print("\n前10行结果:")
# print(df_cash[['party_tp_score', 'party_tp_score_z', 'npvi', 'npvi_z', 
#               'score_diff_abs', 'party_score_pl', 'party_score_ne']].head(10))

# 清理临时变量
df_cash = df_cash.drop(['party_tp_score_z', 'npvi_z', 'score_diff_abs'], axis=1)


# 显示结果
# print("party_tp_score_z分布:")
# print(f"大于0的样本数: {(df_cash['party_tp_score_z'] > 0).sum()}")
# print(f"小于0的样本数: {(df_cash['party_tp_score_z'] < 0).sum()}")
# print(f"等于0的样本数: {(df_cash['party_tp_score_z'] == 0).sum()}")

# print("\n新特征的基本统计:")
# print("party_score_pl描述统计:")
# print(df_cash['party_score_pl'].describe())
# print("\nparty_score_ne描述统计:")
# print(df_cash['party_score_ne'].describe())

# # 查看前10行结果
# print("\n前10行结果:")
# print(df_cash[['party_tp_score', 'party_tp_score_z', 'party_score_pl', 'party_score_ne']].head(10))

# 清理临时变量
# df_cash = df_cash.drop(['party_tp_score_z', 'npvi_z'], axis=1)
```
